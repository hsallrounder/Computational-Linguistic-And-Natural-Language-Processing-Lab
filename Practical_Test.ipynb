```python
from nltk.tokenize import word_tokenize
from nltk.util import bigrams
```

```python
d1="Jack London traveled to Oakland"
d2="Jack London traveled to the city of Oakland"
d3="Jack traveled from Oakland to London"

# doc to Bigrams
def document_to_bigram(doc):
  return set(bigrams(word_tokenize(doc)))

# Jaccard Coefficient
def Jaccard_Coeff(d1,d2):
  bgram_d1=document_to_bigram(d1)
  bgram_d2=document_to_bigram(d2)
  return len(bgram_d1.intersection(bgram_d2))/len(bgram_d1.union(bgram_d2))

print("J(d1, d2) =",Jaccard_Coeff(d1,d2))
print("J(d1, d3) =",Jaccard_Coeff(d1,d3))
```
